## Problem Statement
A bank’s compliance team performs Know Your Customer (KYC) and Anti-Money Laundering (AML) checks for every new customer.

Currently, the process is mostly manual:

Officers review identity documents

Analysts examine customer profiles

Risk teams assign scores

Reports are written manually

This leads to:

Slow processing time

Inconsistent risk assessments

Human errors

High operational workload

The bank wants to explore whether an Agentic AI system can assist by automating parts of the compliance review.

```python
# ── Install ──────────────────────────────────────────────
!pip install -q crewai

# ── Imports ──────────────────────────────────────────────
import os
from crewai import Agent, Task, Crew, Process, LLM

os.environ["OPENAI_API_KEY"] = input("Enter your OpenAI API key: ").strip()

# ── LLM & Memory ─────────────────────────────────────────
llm = LLM(model="openai/gpt-4o-mini")
case_memory = []

# ── Agents ───────────────────────────────────────────────
supervisor = Agent(
    role="KYC/AML Supervisor",
    goal="Give high-level risk guidance based on customer data and history.",
    backstory="Senior bank compliance supervisor with 15 years of experience.",
    llm=llm,
    verbose=False
)

collector = Agent(
    role="KYC Document Collector",
    goal="Extract and summarize identity details and documents from customer data.",
    backstory="Specialist in reading and organizing customer KYC information.",
    llm=llm,
    verbose=False
)

validator = Agent(
    role="Risk & AML Validator",
    goal="Assign a numeric risk score and classification based on strict scoring rules.",
    backstory="AML expert who follows precise scoring methodologies.",
    llm=llm,
    verbose=False
)

writer = Agent(
    role="Compliance Summary Writer",
    goal="Create a clear, structured final compliance summary report.",
    backstory="Experienced compliance officer writer who produces officer-ready reports.",
    llm=llm,
    verbose=False
)

# ── Engine ───────────────────────────────────────────────
def kyc_engine(customer_data):
    history = "\n\n".join(f"Case {i+1}:\n{c}" for i, c in enumerate(case_memory[-3:])) or "No previous history."

    tasks = [
        Task(
            agent=supervisor,
            expected_output="3–5 bullet points of supervisor guidance on key risk aspects.",
            description=(
                f"Customer data:\n{customer_data}\n\nPrevious cases:\n{history}\n\n"
                "Give 3–5 bullet points on the main risk aspects to focus on for this customer."
            )
        ),

        Task(
            agent=collector,
            expected_output="Structured summary of documents present, key identity details, and any missing information.",
            description=(
                f"Customer data:\n{customer_data}\n\n"
                "Summarize: what documents are present, key identity details, and any missing information."
            )
        ),

        Task(
            agent=validator,
            expected_output="Risk Score (0–100), Risk Level (Low/Medium/High), and bullet-point reasons.",
            description=(
                f"Customer data:\n{customer_data}\n\n"
                "Calculate the risk score starting at 30. Apply:\n"
                "  +30 if from a sanctioned or high-risk country\n"
                "  +20 if customer is a PEP (Politically Exposed Person)\n"
                "  +10 if documents are missing or incomplete\n"
                "  +20 if there are suspicious transactions\n"
                "  Cap the final score between 0 and 100.\n"
                "  Low: score < 40 | Medium: 40–69 | High: 70–100\n\n"
                "Output EXACTLY in this format:\n"
                "Risk Score: <number>\n"
                "Risk Level: <Low/Medium/High>\n"
                "Reasons:\n- <reason 1>\n- <reason 2>"
            )
        ),

        Task(
            agent=writer,
            expected_output="A complete compliance report with all required sections.",
            description=(
                "Using all outputs from the previous agents, write a professional compliance report with these sections:\n"
                "1. Customer Overview\n"
                "2. KYC Document Details\n"
                "3. Risk Score & Risk Level\n"
                "4. AML Observations\n"
                "5. Recommendation: Approve / Reject / Manual Review (with justification)"
            )
        ),
    ]

    crew = Crew(
        agents=[supervisor, collector, validator, writer],
        tasks=tasks,
        process=Process.sequential,
        verbose=False
    )

    result = crew.kickoff()
    case_memory.append(customer_data[:300])
    return result.raw if hasattr(result, "raw") else str(result)

# ── Run ───────────────────────────────────────────────────
customer_data = """
Customer: Arjun Mehta
Country: High risk country
Documents: Only National ID
Purpose: Open business account
Transactions: Large unexplained international incoming payments
Sanctions: No
PEP: Yes
"""

print("⏳ Running compliance check...\n")
report = kyc_engine(customer_data)
print("=" * 60)
print(report)
print("=" * 60)

print("\n🗂️ Case Memory:")
print("-" * 40)
for i, c in enumerate(case_memory):
    print(f"\nCase {i+1}:\n{c}")
print("-" * 40)
```

## Problem Statement
In today’s fragmented travel ecosystem, users face significant challenges in searching, comparing, and booking travel services (flights, hotels, activities) across multiple platforms, leading to poor user experience, inefficiencies, and low trust in booking processes.

There is a need for a unified digital platform that enables seamless travel planning with integrated search, secure booking, payment processing, and post-booking support, while also delivering personalized recommendations to improve customer engagement.

However, building such a platform within strict constraints of budget (₹30 lakhs) and timeline (6 months), while ensuring high reliability, security, regulatory compliance, and scalability, poses significant challenges. Additionally, dependencies on third-party APIs, lack of defined stakeholder alignment, and incomplete non-functional requirements increase the risk of delivery delays and suboptimal user adoption.

```python
# ============================================================
# STEP 1: Install required libraries
# ============================================================
!pip install -q crewai pypdf python-docx

# ============================================================
# STEP 2: Import required libraries
# ============================================================
import os
from getpass import getpass
from google.colab import files
from pypdf import PdfReader
from docx import Document
from crewai import Agent, Task, Crew, Process, LLM

# ============================================================
# STEP 3: Set your OpenAI API key
# ============================================================
openai_api_key = getpass("Enter your OpenAI API Key: ")
os.environ["OPENAI_API_KEY"] = openai_api_key

# Optional but helpful
os.environ["OPENAI_MODEL_NAME"] = "gpt-4o-mini"

# ============================================================
# STEP 4: Upload file from your computer
# Supports: .txt, .pdf, .docx
# ============================================================
uploaded = files.upload()

# ============================================================
# STEP 5: Function to read uploaded file
# ============================================================
def read_uploaded_file():
    uploaded_files = list(uploaded.keys())

    if not uploaded_files:
        raise ValueError("No file uploaded.")

    file_name = uploaded_files[0]
    print(f"Uploaded file: {file_name}")

    # TXT
    if file_name.lower().endswith(".txt"):
        with open(file_name, "r", encoding="utf-8") as f:
            return f.read()

    # PDF
    elif file_name.lower().endswith(".pdf"):
        reader = PdfReader(file_name)
        text = ""
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
        return text

    # DOCX
    elif file_name.lower().endswith(".docx"):
        doc = Document(file_name)
        text = ""
        for para in doc.paragraphs:
            if para.text.strip():
                text += para.text + "\n"
        return text

    else:
        raise ValueError("Only .txt, .pdf, and .docx files are supported.")

# ============================================================
# STEP 6: Extract text from uploaded document
# ============================================================
requirement_text = read_uploaded_file()

print("\n========== FIRST PART OF DOCUMENT ==========\n")
print(requirement_text[:2000])

# Keep document size manageable
requirement_text = requirement_text[:12000]

# ============================================================
# STEP 7: Create CrewAI LLM
# ============================================================
llm = LLM(
    model="gpt-4o-mini",
    api_key=openai_api_key,
    temperature=0.3
)

# ============================================================
# STEP 8: Define Agents
# ============================================================
business_analyst = Agent(
    role="Enterprise Business Analyst",
    goal="Understand business requirements clearly and convert them into structured insights.",
    backstory=(
        "You are an experienced business analyst in a large enterprise. "
        "You are skilled at understanding BRDs, extracting business goals, "
        "identifying functional requirements, and highlighting important constraints."
    ),
    llm=llm,
    verbose=True
)

risk_reviewer = Agent(
    role="Enterprise Risk Reviewer",
    goal="Identify project, delivery, security, compliance, and adoption risks in the requirement.",
    backstory=(
        "You are a senior enterprise risk reviewer. "
        "You carefully examine requirement documents and identify delivery risks, "
        "security/privacy concerns, hidden assumptions, and missing details."
    ),
    llm=llm,
    verbose=True
)

executive_writer = Agent(
    role="Executive Summary Writer",
    goal="Prepare a crisp, professional, leadership-ready summary for senior stakeholders.",
    backstory=(
        "You are an expert in executive communication. "
        "You write concise, professional summaries for directors, CTOs, and senior leadership."
    ),
    llm=llm,
    verbose=True
)

# ============================================================
# STEP 9: Define Tasks
# ============================================================
task_1 = Task(
    description=f"""
    Read the uploaded business requirement document and extract:

    1. Business objective
    2. Key features requested
    3. Expected business outcomes
    4. Important constraints
    5. Major stakeholders if mentioned

    Requirement document:
    {requirement_text}
    """,
    expected_output="""
    A structured requirement summary with:
    - Business objective
    - Key features
    - Expected outcomes
    - Constraints
    - Stakeholders
    """,
    agent=business_analyst
)

task_2 = Task(
    description=f"""
    Review the uploaded requirement document and identify:

    1. Delivery risks
    2. Security/privacy risks
    3. Adoption/change management risks
    4. Compliance or governance concerns
    5. Missing details or assumptions

    Requirement document:
    {requirement_text}
    """,
    expected_output="""
    A structured enterprise risk report with:
    - Delivery risks
    - Security/privacy risks
    - Adoption risks
    - Compliance concerns
    - Missing details / assumptions
    """,
    agent=risk_reviewer
)

task_3 = Task(
    description="""
    Using the outputs from the Business Analyst and Risk Reviewer,
    prepare a professional leadership-ready summary.

    The summary must include:
    1. Initiative overview
    2. Business value
    3. Key risks
    4. Recommended next steps

    Keep it crisp, clear, professional, and suitable for leadership review.
    """,
    expected_output="""
    A concise executive summary with:
    - Initiative overview
    - Business value
    - Key risks
    - Recommended next steps
    """,
    agent=executive_writer
)

# ============================================================
# STEP 10: Create Crew
# ============================================================
crew = Crew(
    agents=[business_analyst, risk_reviewer, executive_writer],
    tasks=[task_1, task_2, task_3],
    process=Process.sequential,
    verbose=True,
    memory=False
)

# ============================================================
# STEP 11: Run Crew
# ============================================================
result = crew.kickoff()

# ============================================================
# STEP 12: Final Output
# ============================================================
print("\n\n========== FINAL OUTPUT ==========\n")
print(result)
```
